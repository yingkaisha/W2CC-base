## CESM ensemble mean

```python
years = np.arange(1976, 2020)
```

```python
for i_year, year in enumerate(years):
    
    f = open('{}CESM_stn_{}.sh'.format(conf['qsub']['qsub_loc'], year), 'w') 
    
    heads = '''#!/bin/bash -l

#PBS -N {}
#PBS -A P48500028
#PBS -l walltime=6:59:59
#PBS -l select=1:ncpus=12:mem=64GB
#PBS -q casper
#PBS -o {}.log
#PBS -e {}.err

conda activate credit
cd {}
python STN_00_CESM.py {}
'''.format(year, year, year, 
           conf['qsub']['scripts_loc'], 
           year)
    
    print(heads, file=f)    
    f.close()

f = open('{}batch_CESM_stn.sh'.format(conf['qsub']['qsub_loc']), 'w')
for i_year, year in enumerate(years):
    print('qsub CESM_stn_{}.sh'.format(year), file=f)
f.close()
```
